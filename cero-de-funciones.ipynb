## Ceros de funciones
### Método de la bisección

```python
import numpy as np
import matplotlib.pyplot as plt
```

crear una función cualquiera

```python
def f(x):
    return x - np.exp(-x)
```

crear una lista cualquiera de valores de x


```python
x = np.linspace(0, 5, 100)
```

graficar la funcion

```python
plt.plot(x, x)#graficar la identidad
plt.plot(x, np.exp(-x))
```

o se puede graficar lo mismo pero de otra manera:

```python
plt.plot(x, f(x))
plt.hlines(0.0, 0, 5, colors="red")
```

Luego, de que ya conozco el intervalo donde la funcion f(x) = 0, no es necesario graficar
Ahora, reducimos el intervalo de x

```python
x = np.linspace(0, 1, 100)
```

Ahora, usaremos el método de la bisección

```python
a = 0       #valor min del intervalo
b = 1       #valor máx de intervalo

for i in range(5):
   c = 0.5*(a +b)

   condicion = f(a) * f(c)

   if condicion < 0:
    b = c
   else:
    a = c

#asumimos que el ultimo valor de c que calculamos es solución de f(c) = 0
```

Después, graficamos el punto

```python
plt.plot(x, f(x))
plt.hlines(0.0, 0, 5, colors="red")
plt.plot(c, f(c), "o" )
#plt.savefig("cero-funciones-g1.pdf")
```

Notemos que, ya estamos casi cerca del valor donde f(x) = 0. Para acercarnos aún más, debemos aumentar las iteracciones, del calculo de c, en el ciclo for.

```python
a = 0       #valor min del intervalo
b = 1       #valor máx de intervalo

for i in range(10):                #ahora, el rango es hasta 10.
   c = 0.5*(a +b)

   condicion = f(a) * f(c)

   if condicion < 0:
    b = c
   else:
    a = c

plt.plot(x, f(x))
plt.hlines(0.0, 0, 5, colors="red")
plt.plot(c, f(c), "o" )

plt.savefig("cero-funciones-g2.pdf")
```
